```python
mrna_linc_fa="/home/ksuga/biocontainer/ref/mRNA.lncRNA.fa"
def getTranscriptBasicInformation(mrna_lncrna_fa):
	transCategory,transLen,utr3Len,utr5Len,cdsLen,transORF = {},{},{},{},{},{}
	trans_fh = open(mrna_lncrna_fa) 
	for line in trans_fh:
		if line[0]==">":
			a = line.rstrip().split()
			tID = a[0][1:]
			tCat = a[2].split("=")[1]
			transCategory[tID] = tCat
			print()
			if tCat == "mRNA":
				b = a[3].split("=")[1].split("-")
				transORF[tID] = [int(b[0]), int(b[1])+3] 
				cdsLen[tID] = int(b[1]) - int(b[0]) + 3
				utr5Len[tID] = int(b[0]) 
		else:
			line = line.rstrip()
			transLen[tID] = len(line) 
			if tCat == "mRNA":
				utr3Len[tID] = transLen[tID] - transORF[tID][1] 
	trans_fh.close()
	return transCategory,transLen,transORF,utr5Len,utr3Len,cdsLen
```

```python
transCategory,transLen,transORF,utr5Len,utr3Len,cdsLen=getTranscriptBasicInformation(mrna_linc_fa)
```

```python
all_rna_num=0
all_linc_num=0
rna_num=0
linc_num=0
for k,v in transCategory.items():
    if v=="mRNA":
        all_rna_num+=1
    elif v=="lincRNA":
        all_linc_num+=1
    transID=k.split("_")[0]
    if ribo_dic.get(transID):
        if v=="mRNA":
            rna_num+=1
        elif v=="lincRNA":
            linc_num+=1
print(all_rna_num)
print(rna_num)
print(linc_num)    
```
